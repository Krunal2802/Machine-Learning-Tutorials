## Algerian Forest Fires Dataset 
Data Set Information:

The dataset includes 244 instances that regroup a data of two regions of Algeria,namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012.
The dataset includes 11 attribues and 1 output attribue (class)
The 244 instances have been classified into fire(138 classes) and not fire (106 classes) classes.

Attribute Information:

1. Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012)
Weather data observations
2. Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
3. RH : Relative Humidity in %: 21 to 90
4. Ws :Wind speed in km/h: 6 to 29
5. Rain: total day in mm: 0 to 16.8
FWI Components
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
8. Drought Code (DC) index from the FWI system: 7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely Fire and not Fire

# Basic EDA, Feature Engineering and analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv", header = 1)
```

```python
df.head()
```

```python
df.info()
```

## Data Cleaning

```python
## add column of saying that which is the region is 
df.loc[:122,"Region"] = 0
df.loc[122:,"Region"] = 1
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
## removing the Null values
df = df.dropna().reset_index(drop = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop = True)
```

```python
df.iloc[[122]]
```

```python
df.columns
## here you find some columns have sapces in thier name 
```

```python
## removing extra spaces in columns name
df.columns = df.columns.str.strip()
df.columns
```

### changes the required columns as integer data type

```python
df[['month','day','year','Temperature', 'RH', 'Ws']] = df[['month','day','year','Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

### changes the other columns as float data type

```python
## this is a generic way to do this -> you should do like this generally
objects = [features for features in df.columns if df[features].dtypes == 'O'] ## here 'O' referes to the 'Object' datatype 
```

```python
for i in objects:
    if i!='Classes': ## here Classes is omited bcoxz it is the output variable
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

### Lets Save the clean dataset

```python
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index = False) # index = False -> here index of every column also consideres as column so we set it to flase so it stores actual main data
```

# MONTHLY ANALYSIS FOR FIRE by REGION

```python
dftemp = df.loc[df['Region'] == 0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x = 'month', hue='Classes', data= df)
plt.ylabel('No of Fires', weight = 'bold')
plt.xlabel('Months', weight = 'bold')
plt.title("Fire Analysis of Bejaia Region", weight = 'bold')
```

```python
dftemp = df.loc[df['Region'] == 1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x = 'month', hue='Classes', data= df)
plt.ylabel('No of Fires', weight = 'bold')
plt.xlabel('Months', weight = 'bold')
plt.title("Fire Analysis of Sidi-Bel Region", weight = 'bold')
```

Its observed that <b>August and September</b> had the <b>most number of forest fires</b> for both regions. And from the above plot of months, we can understand few things

Most of the fires happened in August and very high Fires happened in only 3 months - June, July and August.

<b>Less Fires was on September</b>
