# Isolation Forest Anomaly Detection

```python
import pandas as pd
df = pd.read_csv("healthcare.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.iloc[:,0],df.iloc[:,1])
```

```python
from sklearn.ensemble import IsolationForest
clf = IsolationForest(contamination=0.2)
## means anomaly_score value > 0.2 then tht point is an outlier
clf.fit(df)
pred = clf.predict(df)
```

```python
pred
```

```python
import numpy as np
out_index = np.where(pred<0)
## -1 as outliers
out_index
```

```python
x = df.values
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1])
plt.scatter(x[out_index,0],x[out_index,1],edgecolors="r")
```
