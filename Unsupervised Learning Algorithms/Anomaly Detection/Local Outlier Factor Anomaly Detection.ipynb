```python
from sklearn.datasets import make_blobs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X,y = make_blobs(n_samples=1000,n_features=2,center_box=(-5,5))
```

```python
X,y
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.neighbors import LocalOutlierFactor
lof = LocalOutlierFactor(n_neighbors=7,contamination=0.2)
```

```python
pred = lof.fit_predict(X)
```

```python
pred
```

```python
import numpy as np
out_index = np.where(pred<0)
## -1 as outliers
out_index
```

```python
plt.scatter(X[:,0],X[:,1])
plt.scatter(X[out_index,0],X[out_index,1],edgecolors="r")
```
